```python
import pandas as pd
import os

# ========== 데이터 저장 폴더 생성 ==========
# processed 폴더가 없으면 자동으로 생성
# exist_ok=True: 폴더가 이미 있으면 오류 안 냄
os.makedirs('./data/processed', exist_ok=True)

# ========== 경로 및 지점 설정 ==========
# 원본 데이터가 저장된 폴더 경로
data_raw_path = './data/raw'

# 분석 대상 5개 한강 지점
locations = ['paldang', 'jamsil', 'noryangjin', 'gwangnareu', 'bogwang']

# ========== 데이터 병합 준비 ==========
# 모든 파일의 데이터를 저장할 리스트
all_data = []

# ========== 각 지점의 Excel 파일 읽기 및 병합 ==========
# 5개 지점 폴더를 순회
for location in locations:
    # 각 지점의 폴더 경로 생성
    location_path = os.path.join(data_raw_path, location)
    
    # 해당 폴더의 모든 파일을 순회
    for file in os.listdir(location_path):
        # .xlsx 파일만 처리 (Excel 형식)
        if file.endswith('.xlsx'):
            file_path = os.path.join(location_path, file)
            
            # Excel 파일을 DataFrame으로 읽기
            # header=1: 2번째 행(인덱스 1)을 컬럼명으로 설정
            df = pd.read_excel(file_path, header=1)
            
            # 읽은 데이터를 리스트에 추가
            all_data.append(df)

# ========== 모든 데이터를 하나로 병합 ==========
# 리스트의 모든 DataFrame을 세로로 합치기
# ignore_index=True: 새로운 인덱스 생성 (0부터 시작)
merged_df = pd.concat(all_data, ignore_index=True)

# ========== 병합된 데이터를 CSV로 저장 ==========
# processed 폴더에 CSV 파일로 저장
# index=False: 행 번호(인덱스)를 파일에 포함하지 않음
merged_df.to_csv('./data/processed/hangang_merged.csv', index=False, encoding='utf-8-sig')

# ========== 결과 출력 ==========
print(f"✅ 병합 완료!")
print(f"총 행 수: {len(merged_df)}")
print(f"컬럼: {merged_df.columns.tolist()}")

# 데이터 미리보기 (첫 5행)
print("\n📊 데이터 샘플:")
print(merged_df.head())
```

```python
# ========== 결측치 확인 ==========
# 각 컬럼별 결측치(NaN) 개수 확인
print("=" * 50)
print("📌 결측치(NaN) 개수:")
print("=" * 50)
print(merged_df.isnull().sum())
```

```python
# ========== 기본 통계 ==========
# 수치형 데이터의 기본 통계 정보
print("\n" + "=" * 50)
print("📊 기본 통계 정보:")
print("=" * 50)
print(merged_df.describe())
```

```python
# 지점별 데이터 개수 확인
print("📍 지점별 데이터 개수:")
print(merged_df['측정소명'].value_counts())
print(f"\n전체: {len(merged_df)}행")
```

```python
# 측정 회차별 확인
print("\n📌 측정 회차 확인:")
print(merged_df['회차'].value_counts())
```

```python
# ========== 날짜별 평균값으로 정규화 ==========
# 팔당댐의 다층 측정으로 인한 불균형 해결
# 각 지점별, 날짜별로 평균값만 추출

normalized_df = merged_df.groupby(['측정소명', '년/월/일'])[
    ['DO(㎎/L)', 'BOD(㎎/L)', 'COD(㎎/L)', 'TP(㎎/L)', '수온(℃)', 'TN(㎎/L)', 'SS(㎎/L)']
].mean().reset_index()

print("✅ 정규화 완료!")
print(f"정규화 후 행 수: {len(normalized_df)}")
print(f"\n지점별 데이터:")
print(normalized_df['측정소명'].value_counts())

# 정규화된 데이터 저장
normalized_df.to_csv('./data/processed/hangang_merged_normalized.csv', index=False, encoding='utf-8-sig')
print(f"\n💾 정규화된 데이터 저장 완료: hangang_merged_normalized.csv")
```

```python
# ========== 지점별 데이터 균등화 ==========
# 모든 지점을 동일한 개수(34개)로 통일
# random_state=42: 재현 가능성을 위해 고정

balanced_df = normalized_df.groupby('측정소명').apply(
    lambda x: x.sample(min(34, len(x)), random_state=42)
).reset_index(drop=True)

# ========== 결과 확인 ==========
print("✅ 균등화 완료!")
print("\n📊 균등화된 지점별 데이터:")
print(balanced_df['측정소명'].value_counts())
print(f"\n총 행 수: {len(balanced_df)}")

# ========== 균등화된 데이터 저장 ==========
balanced_df.to_csv('./data/processed/hangang_merged_balanced.csv', index=False, encoding='utf-8-sig')
print(f"\n💾 균등화된 데이터 저장 완료: hangang_merged_balanced.csv")
```

```python
# ========== 결측치 확인 ==========
print("📌 결측치(NaN) 현황:")
print(balanced_df.isnull().sum())

# ========== 결측치 처리 ==========
# 방법: 각 컬럼의 평균값으로 결측치 채우기
final_df = balanced_df.copy()

# 수치형 컬럼만 평균으로 채우기
numeric_cols = ['DO(㎎/L)', 'BOD(㎎/L)', 'COD(㎎/L)', 'TP(㎎/L)', '수온(℃)', 'TN(㎎/L)', 'SS(㎎/L)']
for col in numeric_cols:
    final_df[col].fillna(final_df[col].mean())

print(f"\n✅ 결측치 처리 완료!")
print(f"처리 방법: 평균값으로 채우기")
print(f"최종 행 수: {len(final_df)}행")

# ========== 지점별 데이터 확인 ==========
print("\n📊 지점별 데이터 (처리 후):")
print(final_df['측정소명'].value_counts())

# ========== 최종 데이터 저장 ==========
final_df.to_csv('./data/processed/hangang_final.csv', index=False, encoding='utf-8-sig')
print(f"\n💾 최종 분석 데이터 저장: hangang_final.csv")

# ========== 최종 통계 ==========
print("\n📈 최종 데이터 통계:")
print(final_df.describe())
```
